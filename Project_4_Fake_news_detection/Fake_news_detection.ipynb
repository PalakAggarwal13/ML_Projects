```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
import nltk
nltk.download('stopwords')
```

```python
import zipfile

with zipfile.ZipFile("archive.zip", "r") as zip_ref:
    zip_ref.extractall("news_data")

fake_news = pd.read_csv("news_data/Fake.csv")
true_news = pd.read_csv("news_data/True.csv")
```

```python
fake_news['label'] = 1
true_news['label'] = 0

df = pd.concat([fake_news, true_news], ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.to_csv("combined_news.csv", index=False)

print("Combined dataset shape:", df.shape)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['content'] = df['subject'] + ' ' + df['title']
```

```python
X = df.drop(columns='label' , axis=1)
Y = df['label']
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemming_content = re.sub('[^a-zA-Z]',' ',content)
  stemming_content = stemming_content.lower()
  stemming_content = stemming_content.split()
  stemming_content = [port_stem.stem(word) for word in stemming_content if not word in stopwords.words('english')]
  stemming_content = ' '.join(stemming_content)
  return stemming_content
```

```python
df['content'] = df['content'].apply(stemming)
```

```python
X = df['content'].values
Y = df['label'].values
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)

preds = model.predict(X_train)
```

```python
metrics.accuracy_score(Y_train,preds)
```

```python
test_preds = model.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,test_preds)
```

```python
for pred in test_preds:
  if (pred == 0):
    print('Real News')
  else:
    print('Fake News')
```
